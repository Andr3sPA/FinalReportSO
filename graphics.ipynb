```python
import pandas as pd
dfRaid0 = pd.read_csv('raid0.csv')
dfRaid1 = pd.read_csv('raid1.csv')
dfRaid5 = pd.read_csv('raid5.csv')
dfSingle = pd.read_csv('single.csv')
```

```python
import matplotlib.pyplot as plt
raid0Write = dfRaid0['write'].mean()
raid1Write = dfRaid1['write'].mean()
raid5Write = dfRaid5['write'].mean()
singleWrite = dfSingle['write'].mean()
promedios = [raid0Write, raid1Write, raid5Write, singleWrite]
columnas = ['Raid 0', 'Raid 1', 'Raid 5', 'single Disk']

# Crear la gr√°fica de barras
plt.figure(figsize=(10, 6))
plt.bar(columnas, promedios, color=['blue', 'green', 'red', 'purple'])
plt.xlabel('Configuration')
plt.ylabel('Velocity I/O (KB/s)')
plt.title('Average of Write')
plt.show()
```
